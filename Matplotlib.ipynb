```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Basics

```python
#linespace se equally seprated nums (total 11) from 0 t0 5
x = np.linspace(0,5,11)
y = x **3
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.title("Title ka naam yahan")
plt.xlabel("x asis ka naam")
plt.ylabel("y axis ka naam")
```

#Subplots

```python
#subplot(rows,colums,number like matrics  1 2 3 neeche 4 5 6 neeche 7 8 9)
#fir subplot k just neeche plot what we need to plot in that sublot
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

#Object Oriented way

```python
fig = plt.figure(figsize=(20,8),dpi = 100) #2000x800 and dpi matlab depth per inch(high pixel krna hoo matlab modre HD, 100 is default , we can do 500 apni marzi pe)
#axis1 = fig.add_axis([right ki taraf kitna khiskana,auppar ki taraf kitna khiskana,length(%),width(%)])
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.5,0.3])
axis2.plot(x,y)
```

#Types of Plots

```python
plt.scatter(x,y)
```

#Histogram

```python
#random module se sample function import kiya.
from random import sample
#sample = ek function hai jo random unique elements choose karta hai.
#1 se 999 tak me se 100 random unique numbers liye
# unko data me store kiya.
data = sample(range(1,1000),100)
plt.hist(data)
```

#Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
#ax and fig bas naam hai
```
