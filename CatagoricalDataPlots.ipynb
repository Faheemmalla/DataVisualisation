```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Qualtative variable :- jo catagory wise devide ho eg gender(issi ko hum catagorical bolte hai)
df = sns.load_dataset('tips')
df
```

#CountPlot

```python
#hum dhek rahe hai kitne mails and females and and aunmai se kitne smoker hai
sns.countplot(x = df['sex'],hue = df['smoker'])
# x yaane x axis pe sex
```

#BarPlot


```python
#bar plot takes x axis and y axis , aak he pe error dega
sns.barplot(x = df['sex'],y = df['total_bill'])
# bar plot mai y axis pe jaata hai quantative variable(data)
# x pe jata hai catagorical
# y mai ye automically average k basis pe compare kr traha hai
#hum estimater se apni marzi k according ye kr sakte hai ki
#kiss basis pe compare hoga
```

```python
import numpy as np
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
#yahan hum sum pe basis pe compare kr rahe hai
#here total tips given by male is 500 and by female is approx 200
```

#BoxPlot

```python
sns.boxplot(x= 'tip',y = 'day',data = df)
#x pe quantative
# y pe catagorical
```

#boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
#jahan pe ies plot mai zaayda mota wahan zaayda data and jahan pe kam wahan pe kam data
```

#stripplot and swarmplot
Stripplot ek simple scatter jaisa plot hota hai jisme ek variable categorical hota hai, aur saare data points directly dikhte hain. Points kabhi-kabhi overlap ho jaate hain.
Swarmplot bhi almost same hai, lekin isme points automatically adjust ho jaate hain taaki overlap na ho. Isse distribution zyada clear dikhta hai, par bohot zyada data me slow ho sakta hai.

basically Saari cheezu ko aak sath plot krne k liye

```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```
