```python
#Qualtative variable :- jo catagory wise devide ho eg gender(issi ko hum catagorical bolte hai)
#Quantitative Variable :- jo quantity wise divide hoo eg age , salary
#Distributionplots are used to plot quantitave data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
#ye seaborn ka khud ka data set hai , bohut saare hai , humne tips wala data set load kiya
df
```

```python
df['size'].unique()
#iesme 5 unique hai implies ye bhi catagorical
```

#Dist Plot or hist plot (donu same)

```python
#ye jo dabbe bante hai graph pe inhe hum bins bolte hai (ienka bin size hota hai and number of bina )
#values according to bins size mai padhta hai
#kde(kernal density estimate) ye line dalta hai to tell us normal hai kya distribution ya left ki tarag ya right ki taraf
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

#Joint Plot(To compare 2 data ya more same plot mai )

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
# x asis pe total bills, y pe tips , kahan se le rahe hai data (df se le rahe hai), kind iesme plot k liye yahan scatter
# bohut saare kind hote hai like (kde,hex,reg)
# wahi bhai jaise graph daalte hai waise he scatter plot dalta hai like total_bill 16.99 and tip 1.01 wahi pe dot padega , waise he for all points
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
#reg yaane regression plot
```

#Pair Plot(important)

```python
sns.pairplot(df,hue='size',palette='rainbow')
#ye sirf numerical wala load kr hai isseliye sirf total bill and tip ka plot ,
#but english wale ka use krke hum catagorize kr saktre hai (aache se samajne k liye)
#hue se catagory wise hoga  (size wise catagortize kiya and each ka alag alag color)
# sex use krke bhi catagorize kr sakte hai
```
